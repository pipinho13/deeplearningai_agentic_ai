```python
from openai import OpenAI 

client = OpenAI() 
MODEL = "gpt-4.1"

def llm_call(system, user):
    response = client.responses.create(
        model=MODEL,
        temperature=0.7,
        input=f"{system}\n\n{user}",
    )
    return response.output_text

def generate_answer(task): 
    system = "You are a helpful assistant that produces high-quality answers." 
    return llm_call(system, f"Task:\n{task}")

def critique_answer(task, answer): 

    system = "You are a strict critic. Identify flaws, gaps, and improvements." 
    
    prompt = f""" Task: {task} Answer: {answer} Give detailed critique. Be specific. """ 
    
    return llm_call(system, prompt)


def evaluate_answer(task, answer): 
    
    system = "You are an evaluator that scores answers from 1 to 10." 
    
    prompt = f""" Task: {task} Answer: {answer} Return ONLY a number from 1-10. """ 
    
    score_text = llm_call(system, prompt) 
    
    try:
        return float(score_text.strip()) 
    except: return 0.0


def reflect_answer(task, answer, critique): 
    
    system = "You improve answers using critique." 
    
    prompt = f""" 
    Task: {task} 
    Current Answer: {answer} 
    Critique: {critique} Produce a significantly improved answer. 
    """ 
    return llm_call(system, prompt)


def agentic_workflow(task, max_iters=5, target_score=8.5): 
    answer = generate_answer(task) 

    print('----- Initial Answer ----\n')
    print(answer)
    
    for i in range(max_iters): 
        print(f"\n--- Iteration {i+1} ---") 
        critique = critique_answer(task, answer) 
        score = evaluate_answer(task, answer) 
        print(f"Score: {score}") 
        print(f"Critique: {critique}") 
        print('\n------------------\n')
        if score >= target_score: 
            break 
        answer = reflect_answer(task, answer, critique) 
    return answer
```

```python
task = "Explain word embeddings to a 10-year-old." 

final_answer = agentic_workflow(task, target_score=10) 

print(final_answer)
```
